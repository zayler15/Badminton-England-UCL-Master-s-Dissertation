```python
#the neccesary library 
import pyreadstat

import pandas as pd 
```

```python
# Define the file paths for the .sav and .csv files
input_file = "year1.sav"


data, metadata = pyreadstat.read_sav(input_file)

# Convert the data to a pandas dataframe
df = pd.DataFrame(data)

# Save the data to a CSV file
output_file = "year1.csv"
df.to_csv(output_file, index=False)
```
