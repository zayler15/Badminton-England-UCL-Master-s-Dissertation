```python
#function to set anything less than 0 to NAN as stated in the survey data 
def set_negative_to_nan(dataframe):
    dataframe[dataframe < 0] = np.nan
    return dataframe

# Apply the function to set negative values to NaN for all columns
full = set_negative_to_nan(merged_df)
```

```python
# Separate continuous and categorical columns
continuous_cols = ['Frequency_Racket', 'Duration_Racket','Mins_Badminton_Perweek',
                   'Badminton_Duration','Badminton_12_Momths','Time','IMD']

categorical_cols = [col for col in full.columns if col not in continuous_cols]

# Replace NaNs with 0 in continuous columns
#Based on logic and assumption if the value is nan its likely they didn't play racket sports or badminton etc.
df_continuous = full[continuous_cols].fillna(0)

# Perform mode imputation for categorical columns
mode_imputer = SimpleImputer(strategy='most_frequent')
df_categorical = pd.DataFrame(mode_imputer.fit_transform(full[categorical_cols]), columns=categorical_cols)

# Combine the imputed dataframes
df_imputed = pd.concat([df_continuous, df_categorical], axis=1)
```

### Detecting Outliers

```python
# Set the threshold for outlier detection
iqr_threshold = 1.5

# Calculate the IQR for each column
q1 = df_imputed.quantile(0.25)
q3 = df_imputed.quantile(0.75)
iqr = q3 - q1

# Define the lower and upper bounds for outliers
lower_bound = q1 - (iqr_threshold * iqr)
upper_bound = q3 + (iqr_threshold * iqr)

# Identify outliers
outliers = ((df_imputed < lower_bound) | (df_imputed > upper_bound)).any(axis=1)

# Print the rows containing outliers
print(df_imputed[outliers])
```

There seem to be a lot of outliers so its more than likely that these are edge cases etc.. so keep them may be important for the analysis.

## Data Transformation

```python
# Function to check the distribution of each variable
def check_distribution(df, variable, alpha=0.05):
    # Checking if the variable is normally distributed
    # H0: Data is drawn from normal distribution
    _, p_value = stats.normaltest(df[variable])
    
    if p_value < alpha:
        # If the null hypothesis can be rejected, the data is not normally distributed
        print(f"Variable {variable} is not normally distributed")
        
        # Check the distribution after log transformation
        transformed, _ = stats.boxcox(df[variable]+1)  # boxcox requires positive values
        _, p_value_transformed = stats.normaltest(transformed)
        
        if p_value_transformed > alpha:
            print(f"Variable {variable} is normally distributed after transformation\n")
        else:
            print(f"Variable {variable} is not normally distributed even after transformation\n")
    else:
        print(f"Variable {variable} is normally distributed\n")

# Iterating over each numeric column in the DataFrame
for column in df_imputed.select_dtypes(include=np.number).columns:
    check_distribution(df_imputed, column)
```

```python
from scipy.stats import boxcox
from sklearn.preprocessing import PowerTransformer

# Function to apply various transformations
def apply_transformations(df, variable, alpha=0.05):
    transformations = {'Log': np.log, 'Square root': np.sqrt, 'Box-Cox': boxcox}
    
    results = []
    
    for transformation_name, transformation_func in transformations.items():
        try:
            transformed_data = transformation_func(df[variable].replace(0, np.nan).dropna()+1)  # replace 0s to prevent errors
            _, p_value = stats.normaltest(transformed_data)
            
            results.append({
                'Variable': variable,
                'Transformation': transformation_name,
                'p_value': p_value,
                'Is_Normal': p_value > alpha,
            })
        except Exception as e:
            results.append({
                'Variable': variable,
                'Transformation': transformation_name,
                'p_value': np.nan,
                'Is_Normal': False,
            })
            
    return pd.DataFrame(results)

# Apply transformations to each numeric column in the DataFrame
results = pd.concat([apply_transformations(df_imputed, column) for column in df_imputed.select_dtypes(include=np.number).columns])

# Print results
print(results)
```

None of the varaibles even after transformation show normal distribution thus may not be correct based on this analysis as fewer people play rackets sports etc... thus transformation of the data might not be neccesary. 

def plot_distribution(dataframe):
    """
    Plot the distribution of all columns in the dataframe.

    Parameters:
    - dataframe: Pandas DataFrame containing the data.

    Returns:
    - None (displays the plots).
    """
    for col in dataframe.columns:
        plt.figure(figsize=(8, 6))
        if dataframe[col].dtype == 'object':
            # Categorical variable, plot bar chart
            value_counts = dataframe[col].value_counts()
            sns.barplot(x=value_counts.index, y=value_counts.values, alpha=0.7)
            plt.xlabel(col)
            plt.ylabel('Frequency')
        else:
            # Numeric variable, plot histogram
            plt.hist(dataframe[col], bins='auto', alpha=0.7, color='blue', edgecolor='black')
            plt.xlabel(col)
            plt.ylabel('Frequency')
        
        plt.title(f'Distribution - {col}')
        plt.show()
        
        
# Plot distribution of all columns
