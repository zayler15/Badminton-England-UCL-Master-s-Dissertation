```python
np.random.seed(111)

# Load your data into variables 'target' and 'features'
target = d0['Badminton']
features = d0.iloc[:, :-2]

# Split the data into training, validation, and test sets
features_train_val0, features_test0, target_train_val0, target_test0 = train_test_split(features, target, test_size=0.2, random_state=42)
features_train0, features_val0, target_train0, target_val0 = train_test_split(features_train_val0, target_train_val0, test_size=0.25, random_state=42)

# Apply SMOTE oversampling to the training set to balance the classes
oversampler = SMOTE(random_state=42)
features_train_resampled0, target_train_resampled0 = oversampler.fit_resample(features_train0, target_train0)

# Apply random undersampling to the resampled training set
undersampler = RandomUnderSampler(random_state=42)
features_train_resampled0, target_train_resampled0 = undersampler.fit_resample(features_train_resampled0, target_train_resampled0)

# Create the model
neural_network0 = Sequential()
neural_network0.add(Dense(389, activation='relu', input_shape=(389,)))
neural_network0.add(Dense(8, activation='relu'))
neural_network0.add(Dropout(0.6))
neural_network0.add(Dense(8, activation='relu'))
neural_network0.add(Dropout(0.6))
neural_network0.add(Dense(8, activation='relu'))
neural_network0.add(Dropout(0.6))
neural_network0.add(Dense(8, activation='relu'))
neural_network0.add(Dropout(0.6))
neural_network0.add(Dense(8, activation='relu'))
neural_network0.add(Dropout(0.6))
neural_network0.add(Dense(8, activation='relu'))
neural_network0.add(Dense(1, activation='sigmoid'))

# Compile the model
neural_network0.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Define early stopping callback
early_stopping_cb = EarlyStopping(monitor='val_loss', patience=10)

# Define the number of folds for cross-validation
n_splits = 5

# Create the stratified k-fold cross-validator
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# Create lists to store evaluation metrics for each fold
fold_losses0 = []
fold_accuracies0 = []

# Convert features_train_resampled to a DataFrame
features_train_resampled0 = pd.DataFrame(features_train_resampled0)

# Perform cross-validation for Neural Network
for fold, (train_index, val_index) in enumerate(skf.split(features_train_resampled0, target_train_resampled0)):
    print(f"Fold {fold + 1}/{n_splits}")

    # Get the training and validation data for this fold
    fold_features_train = features_train_resampled0.iloc[train_index]
    fold_target_train = target_train_resampled0[train_index]
    fold_features_val = features_train_resampled0.iloc[val_index]
    fold_target_val = target_train_resampled0[val_index]

    # Train the model on the training data for this fold
    fold_history0 = neural_network0.fit(fold_features_train, fold_target_train, epochs=150, batch_size=32, validation_data=(fold_features_val, fold_target_val),
                                       callbacks=[early_stopping_cb], verbose=0)

    # Evaluate the model on the validation set for this fold
    fold_loss0, fold_accuracy0 = neural_network0.evaluate(fold_features_val, fold_target_val, verbose=0)

    fold_losses0.append(fold_loss0)
    fold_accuracies0.append(fold_accuracy0)

# Train the final neural network model on the entire resampled training set
history0 = neural_network0.fit(features_train_resampled0, target_train_resampled0, epochs=150, batch_size=32, validation_data=(features_val0, target_val0),
                    callbacks=[early_stopping_cb])

# Define the number of folds for cross-validation for XGBoost
n_splits_xgb = 3

# Create the stratified k-fold cross-validator for XGBoost
skf_xgb = StratifiedKFold(n_splits=n_splits_xgb, shuffle=True, random_state=42)

# Create lists to store evaluation metrics for each fold for XGBoost
fold_accuracies_xgb = []

# Perform cross-validation for XGBoost
for fold, (train_index, val_index) in enumerate(skf_xgb.split(features_train_resampled0, target_train_resampled0)):
    print(f"XGBoost - Fold {fold + 1}/{n_splits_xgb}")

    # Get the training and validation data for this fold
    fold_features_train = features_train_resampled0.iloc[train_index]
    fold_target_train = target_train_resampled0.iloc[train_index]
    fold_features_val = features_train_resampled0.iloc[val_index]
    fold_target_val = target_train_resampled0.iloc[val_index]

    # Train the XGBoost model on the training data for this fold
     # using results from the hyper-parameter tuning
    #in this case due to computation efficiencies it hasn't been run again; the results have instead been saved from the first run 
    xgb_best0 = XGBClassifier(n_estimators=300, 
                              max_depth=7, 
                              learning_rate=0.1, 
                              subsample=0.7, 
                              colsample_bytree=0.7, 
                              eval_metric='logloss', 
                              random_state=42)  # Adjust parameters as needed

    xgb_best0.fit(fold_features_train, fold_target_train)

    # Use the XGBoost model to make probability predictions on the validation set for this fold
    predictions_on_val0XG = xgb_best0.predict_proba(fold_features_val)[:, 1]

    # Calculate and store the accuracy for this fold
    fold_predictions = [1 if pred > 0.5 else 0 for pred in predictions_on_val0XG]
    fold_accuracy = accuracy_score(fold_target_val, fold_predictions)
    fold_accuracies_xgb.append(fold_accuracy)

# Train the final neural network model on the entire resampled training set
history0 = neural_network0.fit(features_train_resampled0, target_train_resampled0, epochs=150, batch_size=32, validation_data=(features_val0, target_val0),
                    callbacks=[early_stopping_cb])

# Train the final XGBoost model on the entire resampled training set
xgb_best0.fit(features_train_resampled0, target_train_resampled0)

# Evaluate the final neural network model on the test set
loss_on_test_nn, accuracy_on_test_nn = neural_network0.evaluate(features_test0, target_test0)
print('Neural network - Test Loss:', loss_on_test_nn)
print('Neural network - Test Accuracy:', accuracy_on_test_nn)

# Evaluate the final XGBoost model on the test set
predictions_on_test_xgb = xgb_best0.predict(features_test0)
accuracy_on_test_xgb = accuracy_score(target_test0, predictions_on_test_xgb)
print('XGBoost - Test Accuracy:', accuracy_on_test_xgb)

# Use the final neural network model to make predictions on the test set
predictions_on_test0_nn = neural_network0.predict(features_test0).flatten()

# Use the final XGBoost model to make predictions on the test set
predictions_on_test0XG = xgb_best0.predict_proba(features_test0)[:, 1]

# Ensemble the predictions
predictions_ensemble0 = (predictions_on_test0_nn + predictions_on_test0XG) / 2
predictions_ensemble0 = [1 if pred > 0.5 else 0 for pred in predictions_ensemble0]

# Evaluate the ensemble model on the test set
accuracy_ensemble0 = accuracy_score(target_test0, predictions_ensemble0)
print('Ensemble - Test Accuracy:', accuracy_ensemble0)
```

```python
# Convert predictions to numpy array
predictions_ensemble0 = np.array(predictions_ensemble0)

# Convert predictions to binary using a threshold (e.g., 0.5)
binary_predictions0 = np.where(predictions_ensemble0 >= 0.5, 1, 0)

# Calculate the confusion matrix
cm = confusion_matrix(target_test0, binary_predictions0)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# Calculate the classification report
report = classification_report(target_test0, binary_predictions0, output_dict=True)

# Access the F1 score, balanced accuracy, precision, and recall from the report
f1_score = report['macro avg']['f1-score']
balanced_accuracy = balanced_accuracy_score(target_test0, binary_predictions0)
precision = report['macro avg']['precision']
recall = report['macro avg']['recall']

# Calculate Brier Score
brier_score = brier_score_loss(target_test0, binary_predictions0.flatten())
print('Brier Score:', brier_score)

# Print the evaluation metrics
print("F1 Score:", f1_score)
print("Balanced Accuracy:", balanced_accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
