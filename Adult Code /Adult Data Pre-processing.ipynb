## Data Transformation

```python
# Function to check the distribution of each variable
def check_distribution(df, variable, alpha=0.05):
    # Checking if the variable is normally distributed
    # H0: Data is drawn from normal distribution
    _, p_value = stats.normaltest(df[variable])
    
    if p_value < alpha:
        # If the null hypothesis can be rejected, the data is not normally distributed
        print(f"Variable {variable} is not normally distributed")
        
        # Check the distribution after log transformation
        transformed, _ = stats.boxcox(df[variable]+1)  # boxcox requires positive values
        _, p_value_transformed = stats.normaltest(transformed)
        
        if p_value_transformed > alpha:
            print(f"Variable {variable} is normally distributed after transformation\n")
        else:
            print(f"Variable {variable} is not normally distributed even after transformation\n")
    else:
        print(f"Variable {variable} is normally distributed\n")

# Iterating over each numeric column in the DataFrame
for column in df_imputed.select_dtypes(include=np.number).columns:
    check_distribution(df_imputed, column)
```

```python
from scipy.stats import boxcox
from sklearn.preprocessing import PowerTransformer

# Function to apply various transformations
def apply_transformations(df, variable, alpha=0.05):
    transformations = {'Log': np.log, 'Square root': np.sqrt, 'Box-Cox': boxcox}
    
    results = []
    
    for transformation_name, transformation_func in transformations.items():
        try:
            transformed_data = transformation_func(df[variable].replace(0, np.nan).dropna()+1)  # replace 0s to prevent errors
            _, p_value = stats.normaltest(transformed_data)
            
            results.append({
                'Variable': variable,
                'Transformation': transformation_name,
                'p_value': p_value,
                'Is_Normal': p_value > alpha,
            })
        except Exception as e:
            results.append({
                'Variable': variable,
                'Transformation': transformation_name,
                'p_value': np.nan,
                'Is_Normal': False,
            })
            
    return pd.DataFrame(results)

# Apply transformations to each numeric column in the DataFrame
results = pd.concat([apply_transformations(df_imputed, column) for column in df_imputed.select_dtypes(include=np.number).columns])

# Print results
print(results)
```

```python
from pandas import DataFrame
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, f_classif
from matplotlib import pyplot

# Prepare input data
def prepare_inputs(X_train, X_test):
    ordinal_features = X_train.select_dtypes(include=['object', 'category']).columns
    numerical_features = X_train.select_dtypes(include=['float64', 'int64']).columns

    ordinal_transformer = OrdinalEncoder()
    numerical_transformer = StandardScaler()

    preprocessor = ColumnTransformer(
        transformers=[
            ('ordinal', ordinal_transformer, ordinal_features),
            ('numerical', numerical_transformer, numerical_features)
        ])

    preprocessor.fit(X_train)
    X_train_enc = preprocessor.transform(X_train)
    X_test_enc = preprocessor.transform(X_test)
    return X_train_enc, X_test_enc

# Prepare target
def prepare_targets(y_train, y_test):
    le = LabelEncoder()
    le.fit(y_train)
    y_train_enc = le.transform(y_train)
    y_test_enc = le.transform(y_test)
    return y_train_enc, y_test_enc

# Feature selection
def select_features(X_train, y_train, X_test):
    fs = SelectKBest(score_func=f_classif, k='all')
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs

# Prepare the dataset
X = C1.iloc[:, :-1]  # Input features
y = C1.iloc[:, -1]  # Target variable

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

# Prepare input data
X_train_enc, X_test_enc = prepare_inputs(X_train, X_test)

# Prepare output data
y_train_enc, y_test_enc = prepare_targets(y_train, y_test)

# Feature selection
X_train_fs, X_test_fs, fs = select_features(X_train_enc, y_train_enc, X_test_enc)

# What are the scores for the features
feature_scores = fs.scores_
column_names = X_train.columns  # Get the column names

for i in range(len(feature_scores)):
    print('Feature %s: %f' % (column_names[i], feature_scores[i]))

# Plot the scores
pyplot.bar(column_names, feature_scores)
pyplot.xticks(rotation=90)
pyplot.show()
```

```python
#finding optimal hyper-parameters
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Define the range of values for eps and min_samples to search over
eps_values = [0.1, 0.5, 0.05]
min_samples_values = [2, 5, 10]

best_score = -1
best_eps = None
best_min_samples = None

# Loop through all combinations of eps and min_samples
for eps in eps_values:
    for min_samples in min_samples_values:
        # Create a DBSCAN instance with the current parameters
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        
        # Fit DBSCAN to the data and get the cluster labels
        try:
            labels = dbscan.fit_predict(C_pca)
            
            # Check if at least two unique labels are present
            if len(set(labels)) >= 2:
                # Evaluate the clustering using silhouette score
                silhouette_avg = silhouette_score(C_pca, labels)
                
                # Evaluate the clustering using Calinski-Harabasz score
                chs_score = calinski_harabasz_score(C_pca, labels)
                
                # Print the scores for the current parameters
                print(f"eps={eps}, min_samples={min_samples}")
                print("Silhouette score:", silhouette_avg)
                print("Calinski-Harabasz score:", chs_score)
                print()
                
                # Update the best parameters if the current scores are better
                if silhouette_avg > best_score:
                    best_score = silhouette_avg
                    best_eps = eps
                    best_min_samples = min_samples
            else:
                print(f"Not enough distinct clusters found for eps={eps}, min_samples={min_samples}")
        except ValueError:
            print(f"DBSCAN failed for eps={eps}, min_samples={min_samples}")

# Print the best parameters found
print("Best parameters:")
print("eps =", best_eps)
print("min_samples =", best_min_samples)
```

```python
#hyper-parameter tuning to find best parameters for the focal loss 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import backend as K
from tensorflow.keras.callbacks import EarlyStopping
from keras_tuner import RandomSearch

# Define Focal Loss function
def focal_loss(gamma=2., alpha=.25):
    def focal_loss_fixed(y_true, y_pred):
        pt_1 = tf.where(tf.equal(y_true, 1), y_pred, tf.ones_like(y_pred))
        pt_0 = tf.where(tf.equal(y_true, 0), y_pred, tf.zeros_like(y_pred))
        return -K.sum(alpha * K.pow(1. - pt_1, gamma) * K.log(K.epsilon()+pt_1))-K.sum((1-alpha) * K.pow(pt_0, gamma) * K.log(1. - pt_0 + K.epsilon()))
    return focal_loss_fixed

def create_model(hp):
    model = keras.Sequential()
    model.add(Dense(388, activation='relu', input_shape=(388,)))
    model.add(Dense(4, activation='relu'))
    model.add(Dropout(0.5)) 
    model.add(Dense(4, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # Hyperparameters to tune
    hp_gamma = hp.Float('gamma', min_value=0.5, max_value=5.0, default=2.0, step=0.1)
    hp_alpha = hp.Float('alpha', min_value=0.1, max_value=0.9, default=0.25, step=0.05)

    model.compile(optimizer='adam', loss=focal_loss(gamma=hp_gamma, alpha=hp_alpha), metrics=['accuracy'])

    return model

# Instantiate the tuner
tuner = RandomSearch(
    create_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=2,
    directory='my_dir',
    project_name='helloworld')

# Early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5)

# Start the search
tuner.search(X_train, y_train,
             epochs=50,
             validation_data=(X_val, y_val),
             callbacks=[early_stopping])

# Get the optimal hyperparameters
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"Best gamma: {best_hps.get('gamma')}, Best alpha: {best_hps.get('alpha')}")
```
