## Outlier Detection - Continous Varaibles

```python

column_name = 'school_year'
column_data = df_child[column_name]

# Compute z-scores for the column
z_scores = (column_data - column_data.mean()) / column_data.std()

# Find outliers where z-score is greater than 6 or less than -6
outliers = column_data[(z_scores > 10) | (z_scores < -10)]

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(column_data.values, showfliers=False)  # Exclude outliers from the plot
plt.scatter(outliers.index, outliers.values, color='red', label='Outliers')
plt.xlabel('Badminton Frequency')
plt.ylabel('Values')
plt.title('Box Plot with Outliers (school_year)')
plt.legend()
plt.show()

# Extract the 'Badminton_Frequency' column
column_name = 'age'
column_data = df_child[column_name]

# Compute z-scores for the column
z_scores = (column_data - column_data.mean()) / column_data.std()

# Find outliers where z-score is greater than 6 or less than -6
outliers = column_data[(z_scores > 10) | (z_scores < -10)]

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(column_data.values, showfliers=False)  # Exclude outliers from the plot
plt.scatter(outliers.index, outliers.values, color='red', label='Outliers')
plt.xlabel('Badminton Frequency')
plt.ylabel('Values')
plt.title('Box Plot with Outliers (age)')
plt.legend()
plt.show()

```

```python


# Assuming you have a DataFrame named 'data' containing your data

# Select the categorical features for multicollinearity test
categorical_features = df_child.select_dtypes(include=['object'])

# Create an empty DataFrame to store the chi-square test results
chi2_results = pd.DataFrame(index=categorical_features.columns, columns=categorical_features.columns)

# Perform chi-square test of independence for each pair of features
for feature1 in categorical_features.columns:
    for feature2 in categorical_features.columns:
        if feature1 != feature2:
            contingency_table = pd.crosstab(df_child[feature1], df_child[feature2])
            chi2, p_value, _, _ = chi2_contingency(contingency_table)
            chi2_results.loc[feature1, feature2] = p_value

# Print the chi-square test results
print(chi2_results)
```

# Data Encoding

```python
#Encoding the data 
X = pd.get_dummies(df_child[[
       'ethnicity', 'school_type', 'urban_rural',
        'Sports_easy', 'Understand_sports_important',
       'Trust_peers', 'School_meals','Gender']],drop_first=True)


X['year'] = df_child['year']
X['age'] = df_child['age']
X['IMD_child'] = df_child['IMD_child']
#Standardising the data 
# Scale the data using StandardScaler
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
#converting the outcome varaible to be binary

df_child['badminton_onceweek'] = df_child['badminton_onceweek'].astype(bool).astype(int)
```

```python
#performing logistic regression to explore the more simple patterns. 
# Separate the features and target variable


# Separate the features and target variable
X = X_scaled
y = df_child['badminton_onceweek']

# Add constant to the features
X = sm.add_constant(X)

# Count the number of samples in each class
class_counts = df_child['badminton_onceweek'].value_counts()

# Calculate inverse probability weights
weights = class_counts.sum() / (class_counts.shape[0] * class_counts)

# Create and fit logistic regression model with inverse probability weights
logit = sm.Logit(y, X)
result = logit.fit(method='bfgs', maxiter=100, disp=False, w=weights[y])

# Print summary of the model
print(result.summary())
```

## Time Based Regression

```python
#with class weights 

# Set pandas display options
pd.set_option('display.max_columns', None)  # Print all columns

# replace 'Time', 'X_scaled', and 'Y' with your actual data
time_col = X_scaled['year']
respondent_id_col = df_child['serial']
feature_col = X_scaled.drop(['year'], axis=1)
target_col = df_child['badminton_onceweek']

# Replace dots and spaces in the column names with underscores
feature_col.columns = feature_col.columns.str.replace('[. ]', '_', regex=True)

# Get the unique times
unique_times = time_col.unique()

# Time-based regression
for time in unique_times:
    time_indices = time_col[time_col == time].index

    X_time = feature_col.loc[time_indices]
    y_time = target_col.loc[time_indices]

    X_train, X_test, y_train, y_test = train_test_split(
        X_time, y_time, test_size=0.2, random_state=42)

    # Calculate class weights
    class_weights = len(y_train) / (2 * np.bincount(y_train))

    # Define and fit the model with class weights
    model = smf.logit("badminton_onceweek ~ " + "+".join(X_train.columns), data=X_train.assign(badminton_onceweek=y_train))
    result = model.fit(method='bfgs', maxiter=100, class_weight={0: class_weights[0], 1: class_weights[1]})

    # Print the model summary for each year
    print(f'Regression results for year {time}:')
    print(result.summary())

    # Use the model to make predictions on the test data
    y_pred = (result.predict(X_test) > 0.5).astype(int)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f'Precision: {precision}, Recall: {recall}, F1-score: {f1}\n\n')
```

## Mixed Effects Model - Time Combined 

```python
X_scaled.columns = X_scaled.columns.str.replace('[. ]', '_', regex=True)

X_train, X_test, y_train, y_test, respondent_id_train, respondent_id_test = train_test_split(
    X_scaled, target_col, respondent_id_col, test_size=0.2, random_state=42)

# Copy the original dataframes before dropping 'age' and 'school_year'
X_train_original = X_train.copy()
X_test_original = X_test.copy()

# Drop 'age' and 'school_year' from the feature set
#X_train = X_train.drop(['age'], axis=1)
#X_test = X_test.drop(['age'], axis=1)
# and a random intercept and random slope for 'time' for each 'respondent_id_train'
model = smf.mixedlm("badminton_onceweek ~ " + "+".join(X_train.columns) + "+ year", 
                    data=X_train_original.assign(badminton_onceweek=y_train), 
                    groups=respondent_id_train, 
                    re_formula="~ 1 + year")
result = model.fit()

# Print the model summary for the combined years
print(f'Regression results for combined years:')
print(result.summary())

# Use the model to make predictions on the test data
y_pred = (result.predict(X_test_original) > 0.5).astype(int)

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Precision: {precision}, Recall: {recall}, F1-score: {f1}\n\n')
```
