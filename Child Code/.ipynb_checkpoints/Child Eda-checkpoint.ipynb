```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
df_child = pd.read_csv('child_df.csv')
```

```python
df_child.columns
```

```python
df_child.dtypes
```

```python

# List of continuous columns
continuous_cols = ['mins_badminton', 'year', 'survey_year','age']

# Convert variables not in continuous_cols to object data type
for col in df_child.columns:
    if col not in continuous_cols:
        df_child[col] = df_child[col].astype('object')
```

```python
df_child.dtypes
```

```python
df_child = df_child.drop('serial', axis =1)
```

```python
# Count the occurrences of 1's and 0's in the "badminton_onceweek" column
value_counts = df_child['badminton_onceweek'].value_counts(normalize=True) * 100

# Set the figure size
plt.figure(figsize=(10, 15))

# Create a bar chart
plt.bar(value_counts.index, value_counts.values)

# Add labels and title
plt.xlabel("Badminton Once a Week")
plt.ylabel("Percentage")
plt.title("Percentage of Badminton Partcipants (Once a Week)")

# Set the y-axis limit to 100 percent
plt.ylim(0, 100)

# Add y-ticks labels as percentages
plt.gca().set_yticklabels(['{:.0f}%'.format(x) for x in plt.gca().get_yticks()])

# Add x-ticks labels
plt.xticks(value_counts.index, ['No', 'Yes'])

# Show the plot
plt.show()
```

```python
# List of categorical columns
categorical_cols = [ 'ethnicity', 'district', 
                    'school_type', 'urban_rural', 'Sports_easy', 'Understand_sports_important', 
                    'Trust_peers', 'Gender','badminton_onceweek']

# List of numerical columns
numerical_cols = ['mins_badminton', 'year','survey_year',]

# Iterate over each categorical column and create a bar plot
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_child, x=col, hue='badminton_onceweek')
    plt.title(f'Badminton participation by {col}')
    plt.show()

# For numerical columns create a scatter plot with 'mins_badminton'
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df_child, x=col, y='mins_badminton', hue='badminton_onceweek')
    plt.title(f'Minutes of badminton vs {col}')
    plt.show()
```

```python
df_child['School_meals'] = df_child['School_meals'].astype("object")
```

```python
# Grouping the data by age and calculating the average minutes of badminton
df_avg_mins = df_child.groupby('age')['mins_badminton'].mean().reset_index()

# Sorting the DataFrame by age
df_avg_mins.sort_values('age', inplace=True)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df_avg_mins['age'], df_avg_mins['mins_badminton'], marker='o')
plt.xlabel('Age')
plt.ylabel('Average Minutes of Badminton')
plt.title('Average Minutes of Badminton by Age')
plt.xlim(5)  
plt.xticks(df_avg_mins['age'][df_avg_mins['age'] >= 5])  
plt.show()
```

```python
from matplotlib.patches import Patch

# Define a custom color palette
custom_palette = ['#FF7F0E', '#1F77B4']

# Define binary labels for school_type
labels = ['Maintained', 'Academy', 'Independent']

# Calculate the proportions
proportions = df_child.groupby(['school_type', 'badminton_onceweek']).size().groupby(level=0).apply(lambda x: 100 * x / x.sum()).reset_index()

# Plotting the proportion of badminton_onceweek by school_type using Seaborn
plt.figure(figsize=(15, 10))
sns.barplot(data=proportions, x='school_type', y=0, hue='badminton_onceweek', palette=custom_palette)

plt.xlabel('School Type')
plt.ylabel('Proportion (%)')
plt.title('Proportion of Badminton Participants by School Type')

# Set the x-axis tick labels using the 'labels' list
plt.xticks(ticks=[0, 1, 2], labels=labels)

# Create custom legend handles and labels
legend_elements = [
    Patch(facecolor=custom_palette[0], label='No'),
    Patch(facecolor=custom_palette[1], label='Yes')
]
plt.legend(handles=legend_elements, title='Badminton Once a Week', loc='upper right')

plt.show()
```

```python
# Define a custom color palette
custom_palette = ['#FF7F0E', '#1F77B4']

# Define binary labels for school_type
labels = ['Maintained', 'Academy', 'Independent']

# Plotting the count of badminton_onceweek by school_type using Seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=df_child, x='school_type', hue='badminton_onceweek', palette=custom_palette)

plt.xlabel('School Type')
plt.ylabel('Count')
plt.title('Badminton Once a Week by School Type')
plt.legend(title='Badminton Once a Week', labels=['No', 'Yes'])

# Set the x-axis tick labels using the 'labels' list
plt.xticks(ticks=[0, 1, 2], labels=labels)

plt.show()
```

```python
# Filter the DataFrame for badminton_onceweek values equal to 1
df_badminton_onceweek_1 = df_child[df_child['badminton_onceweek'] == 1]

# Plotting the histogram of IMD_child for badminton_onceweek values equal to 1
plt.figure(figsize=(10, 6))
sns.histplot(data=df_badminton_onceweek_1, x='IMD_child', bins=10)
plt.xlabel('Index of Multiple Deprivation — Child')
plt.ylabel('Count')
plt.title('Distribution of Badminton Participants by Index of Multiple Deprivation')
plt.show()
```

```python
custom_palette = sns.color_palette("Set2")


# Define binary labels for school_type
labels = ['Urban', 'Rural']

# Plotting the count of badminton_onceweek by school_type using Seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=df_child, x='urban_rural', hue='badminton_onceweek', palette=custom_palette)

plt.xlabel('Area')
plt.ylabel('Count')
plt.title('Badminton Once a Week by Area Type')
plt.legend(title='Badminton Once a Week', labels=['No', 'Yes'])

# Set the x-axis tick labels using the 'labels' list
plt.xticks(ticks=[0, 1], labels=labels)

plt.show()
```

```python
# Group the data by 'School_meals' and count the occurrences of 1
counts = df_child[df_child['badminton_onceweek'] == 1]['School_meals'].value_counts()

# Define binary labels for school meals
labels = ['No Free School Meals', 'Free School Meals']

# Define less bright colors
colors = ['salmon', 'lightblue']

# Plot the counts with less bright colors for bars using Seaborn
sns.barplot(x=labels, y=counts.values, palette=colors)

plt.xlabel('School Meals')
plt.ylabel('Count of Badminton Participants')
plt.title('Badminton Participation (Once a Week) by Free School Meals')
plt.show()
```

```python
# Grouping data by 'school_year' and 'age' to find the average 'mins_badminton'
# Grouping data by 'school_year' and 'age' to find the average 'mins_badminton'
df_grouped_school_year = df_child.groupby('school_year')['mins_badminton'].mean().reset_index()

# Plotting the average 'mins_badminton' by 'school_year' using lineplot
plt.figure(figsize=(10,6))
sns.lineplot(x='school_year', y='mins_badminton', data=df_grouped_school_year)
plt.title('Average minutes of badminton by school year')
plt.ylabel('Average minutes of badminton')
plt.xlabel('School Year')
plt.show()

```

```python
# Filter data for 'badminton_onceweek' equal to 1 and group by 'district' to get counts
# Filter data for 'badminton_onceweek' equal to 1 and group by 'district' to get counts
df_badminton_onceweek = df_child[df_child['badminton_onceweek'] == 1].groupby('district').size().reset_index(name='counts')

# Ensure 'counts' is a numeric data type
df_badminton_onceweek['counts'] = pd.to_numeric(df_badminton_onceweek['counts'])

# Sort the DataFrame by 'counts' in descending order
df_badminton_onceweek = df_badminton_onceweek.sort_values('counts', ascending=False)

# Plot the counts by 'district' with a fixed color
plt.figure(figsize=(10, 6))
sns.barplot(x='district', y='counts', data=df_badminton_onceweek, color='b', order=df_badminton_onceweek['district'])
plt.title('Number of "1"s in "badminton_onceweek" by District')
plt.xlabel('District')
plt.ylabel('Counts')
plt.xticks(rotation=90)  # Optional: rotate x-axis labels if they are overlapping
plt.show()
```
